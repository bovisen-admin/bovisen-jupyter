# Bovisen JupyterLab — First-Launch Setup

Run this notebook **once on first launch** to create the standard working directory structure under `/home/jovyan/work`.

The directories are created with `exist_ok=True` so it is safe to re-run at any time — existing files will not be touched.

```python
import os

BASE = "/home/jovyan/work"

dirs = [
    "notebooks",
    "data",
    "models",
    "exports",
]

for d in dirs:
    path = os.path.join(BASE, d)
    os.makedirs(path, exist_ok=True)
```

```python
print("Setup complete")
print()
print("Directories under", BASE)
for d in sorted(os.listdir(BASE)):
    print(" ", d)
```
